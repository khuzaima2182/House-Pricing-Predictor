# Loading The Dataset

```python
import pandas as pd
df = pd.read_excel('./Pumpkin_Seeds_Dataset.xlsx', sheet_name='Pumpkin_Seeds_Dataset')
print(df[['Area', 'Perimeter']])
```

# Exploring the dataset

```python
mean_columns = df[['Area', 'Perimeter']].mean()
print(mean_columns)
```

```python
median_columns = df[['Area', 'Perimeter']].median()
print(median_columns)
```

```python
std_columns = df[['Area', 'Perimeter']].std()
print(std_columns)
```

# Plotting a Scatter plot to visualize the relation between Area and Perimeter

```python
import matplotlib.pyplot as plt
x = df["Perimeter"]
y = df["Area"]
plt.scatter(x,y, s = 10)
plt.xlabel("Perimeter")
plt.ylabel("Area")
plt.show()
```

# Plotting Histogram to show distribution of Area Column

```python
plt.hist(y)
```

# Splitting the datasets into test and training

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
x = np.array(x).reshape(-1,1)
y = np.array(y).reshape(-1,1)
print(x.shape)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2,random_state = 0)
```

# Fitting a linear Regression model into training data

```python
result = LinearRegression()
result.fit(X_train,y_train)
```

# Predicting the Area of testing data

```python
y_pred = result.predict(X_test)
```

# Plotting the regression line

```python
plt.scatter(X_test,y_test, s = 10)
plt.xlabel("Perimeter")
plt.ylabel("Area")
plt.plot(X_test,y_pred, color = "r")
plt.show()
```
